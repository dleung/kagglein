```python
import pandas
import numpy as np
import sklearn
from sklearn_pandas import DataFrameMapper, cross_val_score
from sklearn import linear_model
```

```python
df_orig = pandas.read_csv('/Users/dleung/www/mlcompetition/training.tsv', sep='\t')
df_orig.columns = ['age', 'workclass', 'fnlwgt', 'education', 'educationNum', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'country', 'income']
df_orig
```

```python
df_test = pandas.read_csv('/Users/dleung/www/mlcompetition/test.tsv', sep='\t')
df_test.columns = ['age', 'workclass', 'fnlwgt', 'education', 'educationNum', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'country']
df_test
```

```python
# Set the Validation matrix
y = np.where(df_orig.income == "<=50K", 0, 1)
```

```python
# Get the categorical features
scalar_columns = ['age','hours-per-week']

# Encode categorical features with one-hot encoding: http://scikit-learn.org/stable/modules/preprocessing.html#encoding-categorical-features
# We want to use the encoder instead of dummies to reuse it for the training set
mapper = DataFrameMapper([
     ('relationship', sklearn.preprocessing.LabelBinarizer()),
     ('sex', sklearn.preprocessing.LabelBinarizer()),
     ('workclass', sklearn.preprocessing.LabelBinarizer()),
     ('education', sklearn.preprocessing.LabelBinarizer()),
     ('marital-status', sklearn.preprocessing.LabelBinarizer()),
     ('occupation', sklearn.preprocessing.LabelBinarizer()),
     ('country', sklearn.preprocessing.LabelBinarizer()),
     ('age', sklearn.preprocessing.StandardScaler()),
     ('hours-per-week', sklearn.preprocessing.StandardScaler()),
])
# Fit the category transformer to training data, and output the transform data
training_categorical_df = mapper.fit_transform(df_orig) 

# Using the same category transformer, generate the categorical matrix for the test data
test_categorical_df = mapper.transform(df_test)
```

```python
# Logistic Regression
clf = linear_model.LogisticRegression(C=1.0, penalty='l1', tol=1e-6)
clf.fit(training_categorical_df, y)  

# Predict test data
clf.predict(test_categorical_df)
```
